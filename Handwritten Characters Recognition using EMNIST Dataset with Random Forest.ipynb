```python
from emnist import list_datasets,extract_training_samples, extract_test_samples
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pandas as pd
```

# import data test dan validasi emnist

```python
##training
nomor,label_nomor= extract_training_samples('digits')
huruf,label_huruf= extract_training_samples('letters')
print('ini:',nomor.shape, label_nomor.shape)


#ambil hanya 48000 (yang telah diacak) data angka supaya menyeimbangi data huruf
nomor=np.array(nomor[0:48000,:,:])
label_nomor=np.array(label_nomor[0:48000])

#dicek bahwa data udah random
print(label_nomor[0:20])
fig=plt.figure(figsize=(8,8))
for i in range(0,20):
    fig.add_subplot(8,8,i+1)
    plt.imshow(nomor[i,:,:])
plt.show()

#ambil data huruf untuk a sampai j dengan lbel dengan angka 10-19
huruf=np.array(huruf[np.argwhere(label_huruf<=10),:,:])
huruf=np.squeeze(huruf,axis=1)
label_huruf=np.array(label_huruf[np.argwhere(label_huruf<=10)])
label_huruf=np.squeeze(label_huruf,axis=1)

#validasi
vnomor,label_vnomor= extract_test_samples('digits') 
vhuruf,label_vhuruf= extract_test_samples('letters')
vhuruf=np.array(vhuruf[np.argwhere(label_vhuruf<=10),:,:])
vhuruf=np.squeeze(vhuruf,axis=1)
label_vhuruf=np.array(label_vhuruf[np.argwhere(label_vhuruf<=10)])
label_vhuruf=np.squeeze(label_vhuruf,axis=1)

#mengubah label pada dataset huruf menjadi label huruf
#data training
label_hurufbaru=[]
for number in label_huruf:
          number=number+96
          label_hurufbaru.append(chr(int(number)))
#data validasi
label_vhurufbaru=[]
for number in label_vhuruf:
          number=number+96
          label_vhurufbaru.append(chr(int(number)))
    
#menampilkan gambar training huruf beserta label huruf yang baru    
print(label_hurufbaru[0:20])
fig=plt.figure(figsize=(8,8))
for i in range(0,20):
    fig.add_subplot(8,8,i+1)
    plt.imshow(huruf[i,:,:])
plt.show()
print('ini:',nomor.shape, label_nomor.shape)
#gabung
training=tf.concat([nomor,huruf],axis=0)
label_training=tf.concat([label_nomor,label_hurufbaru],axis=0)
validasi=tf.concat([vnomor,vhuruf],axis=0)
label_validasi=tf.concat([label_vnomor,label_vhurufbaru],axis=0)
```

# training model

```python
#training dengan random forest 
from sklearn.ensemble import RandomForestClassifier

forest_clf = RandomForestClassifier(random_state=42)
forest_clf.fit(training, label_training)
```

# load model (ga perlu training lagi)

```python
# Recreate the exact same model, including its weights and the optimizer
model= tf.keras.models.load_model('model_terakhir_lagi.h5')

# Show the model architecture
#print(model_disimpan.summary())
score = model.evaluate(validasi,label_validasi,verbose=2)
print('Test loss model terbaik:', score[0])
print('Test accuracy model terbaik:', score[1])
```

# import gambar datates

```python
import cv2
from os import listdir
from PIL import Image,ImageOps as IO
from numpy import asarray

def loadImages(path):
    # return array of images
    imagesList = listdir(path)
    loadedImages = []
    for image in imagesList:
        img = Image.open(path + "/" + image)
        img=IO.grayscale(img)
        img=IO.invert(img)
        img=img.resize((28,28))
        img=asarray(img)
        loadedImages.append(img)
    return loadedImages

test=[] #membuat list (kirain array tapi kok masih butuh asarray) menyimpan gambar   
path = "D:/Kuliah/datates"
# your images in an array
imgs = loadImages(path)

for img in imgs:
    test.append(img)
    
test=asarray(test) #harus diubah menjadi bentuk array
np.shape(test)
```

# Prediksi

```python
label_prediksi=np.argmax(model.predict(test),axis=1)

print(label_prediksi[0:20])

fig=plt.figure(figsize=(8,8))
for i in range(0,20):
    fig.add_subplot(8,8,i+1)
    plt.imshow(test[i,:,:])
plt.show()

#hasil prediksi buruk meskipun akurasi model pada data validasi cukup tinggi pada 91%
#hal ini, seperti terlihat di bawah, disebabkan oleh tipisnya tulisan mahasiswa yang menjadi data tes sehingga sulit dideteksi sebagai pixel
```

# t-sne

```python
from sklearn.manifold import TSNE
tsne = TSNE(perplexity=8,early_exaggeration=24,init='random',verbose=1,random_state=42)

trainingreshaped=test.reshape(1576,28*28)
```

```python
# use fit_transform instead of fit, as TSNE has no transform method
tsne_data = tsne.fit_transform(trainingreshaped)
```

```python
import pandas as pd
import seaborn as sn
# creating a new data frame which help us in ploting the result data
tsne_data = np.vstack((tsne_data.T,label_prediksi)).T
tsne_df = pd.DataFrame(data=tsne_data, columns=("dimensi1", "dimensi2", "label"))
# Ploting the result of tsne
plt.figure(figsize=(20,17))
sn.FacetGrid(tsne_df, hue="label", size=9).map(plt.scatter, 'dimensi1', 'dimensi2').add_legend()
plt.title('Visualisasi Data Tes dengan T-SNE')

plt.show()
```
